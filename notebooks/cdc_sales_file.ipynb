```python
import pandas as pd
import random
from datetime import datetime, timedelta

rows = []
operations = ["INSERT", "UPDATE", "DELETE"]

base_time = datetime(2025, 1, 1, 9, 0, 0)

for i in range(1, 1001):
    op = random.choices(
        operations,
        weights=[0.6, 0.3, 0.1]
    )[0]

    valor = round(random.uniform(50, 1000), 2) if op != "DELETE" else None

    rows.append({
        "id": random.randint(1, 300),
        "cliente_id": random.randint(100, 200),
        "produto_id": random.randint(1000, 1100),
        "valor": valor,
        "operation": op,
        "timestamp": (base_time + timedelta(minutes=i)).strftime("%Y-%m-%d %H:%M:%S")
    })

df = pd.DataFrame(rows)

# CAMINHO CORRETO (SEM /dbfs)
file_path = "/Volumes/main_cdc/default/raw/cdc_sales_big.csv"

df.to_csv(file_path, index=False)
```
