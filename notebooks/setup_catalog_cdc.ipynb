```python
%sql

-- Criando Catalog
CREATE CATALOG IF NOT EXISTS main_cdc;
USE CATALOG main_cdc;
```

```python
%sql

-- Criando Schemas
CREATE SCHEMA IF NOT EXISTS main_cdc.bronze;
CREATE SCHEMA IF NOT EXISTS main_cdc.silver;
CREATE SCHEMA IF NOT EXISTS main_cdc.gold;

```

```python
%sql
-- Criando Volumes
CREATE VOLUME IF NOT EXISTS main_cdc.default.raw;
CREATE VOLUME IF NOT EXISTS main_cdc.default.schema;
CREATE VOLUME IF NOT EXISTS main_cdc.default.checkpoints;
```
